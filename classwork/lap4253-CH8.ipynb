```python
#This is the challenge from 2/24, where it tells you if a matrix is diagonally dominant, it wasn't explicitly well named



import numpy as np

A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)

length = A.shape[1]
dominant = True

for i in range(0,length):
    diagonalValue = A[i, i]
    rowSum = np.sum(np.abs(A[i])) - diagonalValue
        
    if (diagonalValue < rowSum):
        dominant = False
    
if (dominant == True):
    print(A)
    print("Is the matrix diagonally dominant? YES!")

if dominant == False:
    print(A)
    print("Is the matrix diagonally dominant? NO!")
#The output will look like this: "is the matrix diagonally dominant? True"
```
