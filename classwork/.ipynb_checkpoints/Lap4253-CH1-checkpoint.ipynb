```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.subplot(2, 2, 1)   # 2 rows, 1 column, first plot
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2)   # second plot
plt.plot(x, np.cos(x))

plt.subplot(2,2, 3)
plt.plot(x, np.tan(x))

plt.subplot(2,2,4)
plt.plot(x, (np.sin(x))*(np.tan(x)))

plt.show()
```
