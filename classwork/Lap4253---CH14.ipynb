```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi # Interval
N = 50 # Number of sample points
# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)
h = (b - a) / (N - 1)

#forward
df_dx_forward = (f(x + h) - f(x))/h

#backward
df_dx_backward = (f(x) - f(x - h))/h

#central
df_de_central = (f(x + h) - f(x - h))/ (2*h)


plt.figure(figsize=(10,10), tight_layout = True)
plt.subplot(2,2,1)
#plot cos(x)
plt.plot(x, np.cos(x), label='Exact Derivative', color = "red")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()


#plot forward difference method
plt.subplot(2,2,2)
plt.plot(x, df_dx_forward, label="Forward difference derivative", color = "yellow")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()


#plot backwards difference method
plt.subplot(2,2,3)
plt.plot(x, df_dx_backward, label="Backwards difference derivative", color = "green")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()


#plot central difference method
plt.subplot(2,2,4)
plt.plot(x, df_dx_backward, label="Backwards difference derivative", color = "purple")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
