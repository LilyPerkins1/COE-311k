```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

def approxPi(numSamples):
    counter = 0

    for s in range (0, int(numSamples)):
        randomX = random.uniform(-1,1)
        randomY = random.uniform(-1,1)
     
        if (math.sqrt(randomX*randomX + randomY*randomY) < 1): ##if radius is less than one, count it
            counter += 1
    
    estPi = 4 * (counter/numSamples)

    return estPi;


absoluteErrorList = []
relativeErrorList = []

for i in range (1, 100):
    newPi = approxPi(1000)
    absoluteError = abs(newPi - np.pi)
    relativeError = abs((newPi - np.pi)/np.pi)
    
    absoluteErrorList.append(absoluteError)
    relativeErrorList.append(relativeError)
    
x = range(1,100)

plt.plot(x, absoluteErrorList, label="Absolute Error")
plt.plot(x, relativeErrorList, label="Relative Error")
plt.legend()
plt.show()
```
