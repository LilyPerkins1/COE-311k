```python
#jacobi
import time 
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
# Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
        print("X is: ", x)
    return x, max_iterations

# Example usage
A = np.array([[10, -1, 2, 0, 0],
              [-1, 11, -1, 3, 0],
              [2, -1, 10, -1, 0],
              [0, 3, -1, 8, -2],
              [0, 0, 0, -2, 9]])

b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 1000
              
start_time = time.perf_counter()
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
end_time = time.perf_counter()
total_time = end_time - start_time

print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print("Time of execution: ")
print(f" {total_time:.5f}", "seconds")
print("\n")
print("Using Jacobi here, we notice it only took 1 iteration of the formulas to identify what the solution to our problem was. It also only took 0.00081 seconds, which is because it only had to do 3 formulas. This clearly converges to the answer, which works because the matrix is well structured for this type of analysis.")
```

```python
#gauss-Seidel
#leads to faster convergence
import numpy as np
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        
# Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
        print("X is: ", x)

    return x, max_iterations

A = np.array([[10, -1, 2, 0, 0],
              [-1, 11, -1, 3, 0],
              [2, -1, 10, -1, 0],
              [0, 3, -1, 8, -2],
              [0, 0, 0, -2, 9]])

b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100

start_time = time.perf_counter()
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)
end_time = time.perf_counter()
total_time = end_time - start_time

print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print("Time of execution: ")
print(f" {total_time:.5f}", "Seconds")
print("\n")
print("Here, we can easily see that the answer converges to the same solution as the jacobi method identified, however, it took 3 thousands of a second longer to do so. While the Gauss-Seidel method tends to be quicker, given the simplicity of the solution, this advantage isn't explicit currently. However, it does converge, and comes to the correct answer.")
```
