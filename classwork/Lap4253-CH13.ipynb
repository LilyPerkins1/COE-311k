```python
import numpy as np
import time
import matplotlib.pyplot as plt

# build matrices and store in a list
M = [
    np.random.randn(3,3), np.random.randn(3,3),
    np.random.randn(6,6), np.random.randn(6,6),
    np.random.randn(12,12), np.random.randn(12,12),
    np.random.randn(24,24), np.random.randn(24,24),
    np.random.randn(48,48), np.random.randn(48,48),
    np.random.randn(96,96), np.random.randn(96,96),
    np.random.randn(192,192), np.random.randn(192,192),
    np.random.randn(384,384), np.random.randn(384,384)
]

# naive multiplication
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                
    return C

sizes = [3,6,12,24,48,96,192,384]
nativeTimes = []
numpyTimes = []
# timing
nativeTimeStart = time.perf_counter()
print("Native")
for i in range(0, len(M), 2):
    nativeTimeStart = time.perf_counter()
    A = M[i].tolist()
    B = M[i+1].tolist()
    C = naive_matrix_multiplication(A, B)
    nativeTimeEnd = time.perf_counter()
    nativeTimes.append(nativeTimeEnd - nativeTimeStart)
    print("Total time:", nativeTimeEnd - nativeTimeStart)
print()
print("numpy")
for i in range(0, len(M), 2):
    numpyTimeStart = time.perf_counter()
    C = M[i] @ M[i+1]
    numpyTimeEnd = time.perf_counter()
    numpyTimes.append(numpyTimeEnd - numpyTimeStart)
    print("Total time:", numpyTimeEnd - numpyTimeStart)
print()
plt.plot(sizes,numpyTimes)
plt.plot(sizes,nativeTimes)
plt.title("Run Times")
plt.ylabel("Time in seconds")
plt.xlabel("size of matrices multiplied")
plt.plot()
print("It's clear that the native function to multiply two matrices has a higher computational cost than the numpy, optimized, operator. For the optimized operator, no matter the size, the run time to multiply the matrices is virtually 0 seconds. However, for the native function, it has a compounding effect; for each larger matrix the computational time is significantly larger.")
```
