```python
import numpy as np
v1 = np.array([1, 2, 3])
v2 = np.array([4,5,6])

print(v1)
print(v2)
```

```python
v_add = v1 + v2
v_sup = v1 - v2

print(v_add, v_sup)
```

```python
scalar = 2
v_scaled = scalar * v1
print(v_scaled)
```

```python
dot_product = np.dot(v1,v2)
print(dot_product)
```

```python
magnitude_v1 = np.linalg.norm(v1)
print(magnitude_v1)
```

```python
#using matrixes to solve a system of equations
A = np.array([[2,3],[4,-1]])
b = np.array([5,3])
print("matrix A\n", A)
print("matrix b\n", b)
x = np.linalg.solve(A,b)
print("Solution: (x, y):", x)
```

```python
#eigenvalues/eigenvectors
A = np.array([[2,1],[1,2]])
eigenvalues, eigenvectors = np.linalg.eig(A) #.eig returns two things, so it needs two inputs
print("eigenvalues:\n", eigenvalues, "\neigenvectors \n", eigenvectors)
```

```python
#inversed matrix
a = np.array([[2,7],[4,6]])
a_inv = np.linalg,inv(a)
print(a)
print(a_inv)
```
