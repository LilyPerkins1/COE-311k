```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
plt.plot(x, np.tan(x))
plt.show()

plt.subplot(1, 1, 1)
plt.plot(x, np.arctan(x), 'r')
plt.show()
plt.subplot(1, 1, 1)
plt.plot(x, np.cos(x), 'g')
plt.show()
plt.subplot(1, 1, 1)
plt.plot(x, np.arccos(x), 'r')
plt.show()
plt.subplot(1, 1, 1)
plt.plot(x, np.arcsin(x), 'g')
plt.show()
```
