```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

##start pi function
def approxPi(numSamples):
    counter = 0

    for s in range (0, int(numSamples)):
        randomX = random.uniform(-1,1)
        randomY = random.uniform(-1,1)
     
        if (math.sqrt(randomX*randomX + randomY*randomY) < 1): ##if radius is less than one, count it
            counter += 1
    
    estPi = 4 * (counter/numSamples)

    return estPi;
##end pi function
```

```python
errorList = []

for i in range (10,10000, 20):
    newPi = approxPi(i)
    errorCalc = round(newPi - math.pi, 15)
    errorMag = abs(errorCalc)
    errorList.append(errorMag)
    
x = list(range(0, 10000, 20))

plt.plot(x, errorList)
plt.show()
```
