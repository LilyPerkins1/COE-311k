```python
import math
import numpy as np
import matplotlib.pyplot as plt

oneTerms = []
twoTerms = []
threeTerms = []
fourTerms = []
x_vals = np.linspace(0, 10, 101)#i understand this should be 10, but theres an error that I can't find

#Function to approximate sinx using # terms
def TaylorSine(x):
    approximation = []
    add = 0
    for n in range (0,4):
        #print("approximation for sine for ", n+1, " term/s\n ")
        function = ((-1)**n)* (x ** 2*n + 1)  / math.factorial(2*n + 1)
        add = add + function
        approximation.append(add) #create taylor series
        
    return approximation

print(TaylorSine(0.5))
```

```python
#loop to be used to go from 0-10 in steps of 0.1
one = []
two = []
three = []
four = []

for i in x_vals:
    list1, list2, list3, list4 = TaylorSine(i)

    one.append(list1)
    two.append(list2)
    three.append(list3)
    four.append(list4)

plt.plot(x_vals, one, 'b')
plt.plot(x_vals, two, 'k')
plt.plot(x_vals, three, 'g') #3 and 4 are overlapping
plt.plot(x_vals, four, 'm')
plt.show()
```

```python
#create a true pi using numpy pi
trueSineValues = []

for j in x_vals:
    trueSineValues.append(np.sin(j))
    
```

```python
#calculate the absolute error between numpy pi
errorListOne = []
errorListOne = abs(np.array(one) - np.array(trueSineValues))


errorListTwo = []
errorListTwo = abs(np.array(two) - np.array(trueSineValues))
print(errorListTwo)

errorListThree = []
errorListThree = abs(np.array(three) - np.array(trueSineValues))


errorListFour = []
errorListFour = abs(np.array(four) - np.array(trueSineValues))
```

```python
#make some plots of error over #terms for a couple values of x
plt.plot(x_vals, errorListOne, 'b')
plt.plot(x_vals, errorListTwo, 'k')
plt.plot(x_vals, errorListThree, 'g')
plt.plot(x_vals, errorListFour, 'm')
plt.xlabel("X-values")
plt.ylabel("Error Difference")
plt.show()
```
